```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math

a=1.#минимальное значение
b=1.#коэффициент сдвига
M=b# математическое ожидание
D=(2/(a**2))#дисперсия
print (M)
print (D)

#сгенерируем выборку 1000
laplace = sts.laplace()#задаем функцию Лапласа
sample = laplace.rvs(size=1000)#сгенерируем случайную выборку 1000 значений
x = np.linspace(1,10,1000)#определяем массив для плотности распределения
pdf = laplace.pdf(x)#вычисляем значения плотности распределения от выборки
#строим гистрограмму выборки и теоретическую функцию распределения
plt.plot(x, pdf, label='функция распределения')#строим функцию распределения
plt.hist(sample, bins =50, range=(1,10), density=True)#строим гистаграмму
plt.ylabel('$f(x)$')#подписываем ось у
plt.xlabel('$x$')#подписываем ось х
plt.legend(loc='best')
```

```python
#сгенерируем выборку 1000 при n=5
n=5
massiv_5=np.array([])#обозначим массив выборочных средних
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    massiv_5i = laplace.rvs(n)#сгенерируем выборку
    sum_5i=sum(massiv_5i)/n#определение выборочного среднего
    massiv_5=np.append(massiv_5,sum_5i)#добавляем в массив новое выборочное среднее
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
M=1.#Математическое ожидание Лапласа
D=2.#дисперсия Лапласа
D_5=D/n#дисперсия для нормального распределения при n=5
s_5=math.sqrt(D_5)#среднеквадратичное отклонение(сигма) для нормального распределения выборки средних
n_5 = sts.norm(M, s_5)#определяем нормальное распределение Лапласа
pdf5 = n_5.pdf(x[:500])#получаем плотность распределения
#строим гистограмму  при n=5 и плотность описывающую массив нормального распределения n=5
plt.plot(x[:500], pdf5, label='функция распределения(n=5)')#строим функцию распределения при n=5
plt.hist( massiv_5, bins =80, range=(1,5), density=True)#строим гистаграмму
plt.ylabel('$f(x)$')#подписываем ось у
plt.xlabel('$x$')#подписываем ось х
plt.legend(loc='best')
```

```python
#сгенерируем выборку 1000 при n=10
n=10
massiv_10=np.array([])#обозначим массив выборочных средних
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    massiv_10i = laplace.rvs(n)#сгенерируем выборку
    sum_10i=sum(massiv_10i)/n#определение выборочного среднего
    massiv_10=np.append(massiv_10,sum_10i)#добавляем в массив новое выборочное среднее
    i+=1
#определим параметры нормального распределения описывающие массив 1000  при n=10
M=1.#Математическое ожидание Лапласа
D=2.#дисперсия Лапласа
D10=D/n#дисперсия для нормального распределения приближающего выборку средних
s_10=math.sqrt(D10)#среднеквадратичное отклонение для нормального распределения выборки средних
norm10 = sts.norm(M, s_10)#определяем нормальное распределение с расчетными характеристиками Лапласа
pdf10 = norm10.pdf(x[:300])#получаем плотность распределения
#строим гистограмму массива средних при n=10 и плотность описывающую этот массив нормального распределения
plt.plot(x[:300], pdf10, label='функция распределения(n=10)')
plt.hist(massiv_10, bins =80, range=(1,3), density=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
#сгенерируем выборку 1000 при n=100
n=100
massiv_100=np.array([])#обозначим массив выборочных средних
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    massiv_100i = laplace.rvs(n)#сгенерируем выборку
    sum_100i=sum(massiv_100i)/n#определение выборочного среднего 
    massiv_100=np.append(massiv_100,sum_100i)#добавляем в массив новое выборочное среднее
    i+=1
#определим параметры нормального распределения описывающие массив 1000 выборочного среднего  при n=100
M=1.#Математическое ожидание Лапласа
D=2.#дисперсия Лапласа
D100=D/n#дисперсия для нормального распределения выборки средних
s_100=math.sqrt(D100)#среднеквадратичное отклонение(сигма) для нормального распределения выборки средних
norm100 = sts.norm(M, s_100)#определяем нормальное распределение с расчетными характеристиками Лапласа
pdf100 = norm100.pdf(x[:300])#плотность распределения
#строим гистограмму массива средних при n=100 и плотность  нормального распределения при n=100
plt.plot(x[:300]-1, pdf100, label='функция распределения(n=100)')#строим функцию распределения при n=5
plt.hist(massiv_100, bins =100, range=(0,5), density=True)#строим гистаграмму
plt.ylabel('$f(x)$')#подписываем ось у
plt.xlabel('$x$')#подписываем ось х
plt.legend(loc='best')
```

## <b>Распределение  Лапласа (или двойное экспоненциальное) распределение имеет функцию плотности распределения:f(x) = 1/(2b)*e-|x-a|/b, 
    где a   - среднее распределения, 
    b   - параметр масштаба, 
    e   - число Эйлера .
    При больших объемах выборки (n>10) ее выборочное среднее стремится к нормальному распределению. Нормальным распределением хорошо описываются распределение выборочного среднего.</b>
